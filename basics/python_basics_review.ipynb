```python
class Human: 
    """
    This is a class that represents a very simplified human.
    """
    def __init__(self, age, name):
        self._age = age
        self._name = name
        
    def __str__(self):
        return "A human with name " + self._name + "." + " Their age is " + str(self._age) + "."
    
    def __repr__(self):
        return "A human with name " + self._name + "." + " Their age is " + str(self._age) + "."
    
    def older_younger_than(self, age):
        if self._age > age:
            print("Our age is bigger than their age.")
        elif self._age == age:
            print("Our age is equal to their age.")
        else:
            print("Our age is less than their age.")
            
        
        
h = Human(age=25, name="Dusan")

h

# self is implicit thing, human they are talking about
```

```python
h._age
```

```python
h._name
```

```python
h.__str__()
```

```python
h.older_younger_than(45)
```

```python
h.older_younger_than(23)
```
